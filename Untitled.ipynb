# Clasificador bayesiano

```python
def getWiByClass(X,y):
    w0_mask = y==0
    w1_mask = y==1
    SS.fit(X)
    w0      = SS.transform(X[w0_mask])
    w1      = SS.transform(X[w1_mask])
    return SS,[w0,w1]

def getParameters(i):
    X , y   = Xys_train[i]
    _SS,(w0,w1) = getWiByClass(X,y)
    w0_mean = w0.mean(axis=0)
    w1_mean = w1.mean(axis=0)
    w0_cov  = np.cov(w0.T,bias=False)
    w1_cov  = np.cov(w1.T,bias=False)
    D       = X.shape[1]
    MEANS   = [w0_mean,w1_mean]
    COVS    = [w0_cov,w1_cov]
    Ns      = np.array([w0.shape[0],w1.shape[0]])
    return _SS,MEANS,COVS,Ns
# _SS,MEANS,COVS,Ns = getParameters(0)
```

```python
def px(x,mean,cov):
    xMean       =(x-mean)
    covInv      = LA.inv(cov)
    covDet      = LA.det(cov)
    exp         = (-0.5*xMean.T@covInv)@(xMean)
    denominator = np.power(2*np.pi,D/2)*np.sqrt(covDet)
    return (1/denominator)*exp
def classify(i,X,clf):
    _SS,MEANS,COVS,Ns = getParameters(i)
    _X = _SS.transform(X)
    results = []
    for xi in _X:
        results.append(clf(xi,MEANS,COVS,Ns= Ns))
    return np.array(results)
```

```python
def bayesClassifier(x,MEANS,COVS,**kwargs):
    results = []
    for wi_mean,wi_cov in zip(MEANS,COVS):
        results.append(px(x,wi_mean,wi_cov))
    results = np.array(results)
    return np.argmax(results)
```

## Clasificador de minima distancia Euclidiana

```python
def med_px(xi,MEANS,COVS,**kwargs):
    results = []
    for wi_mean,wi_cov in zip(MEANS,COVS):
        xMean = xi-wi_mean
        px = np.sqrt(xMean.T@xMean)
        results.append(px)
    return np.argmin(np.array(results))        
```

## Clasificador de minima distancia Mahalanobis

```python
def mmd_px(xi,MEANS,COVS,**kwargs):
    results = []
    Ns = kwargs.get("Ns") -1 
    _sigma = 0
    for  i,ni in enumerate(Ns):
        _sigma+=ni*COVS[i]
    _sigma = (1/(Ns.sum()))*_sigma
    for wi_mean,wi_cov in zip(MEANS,COVS):
        xMean = xi-wi_mean
        nis =(Ns[0]-1)+(Ns[1]-1)
#         sigma = ((1 / nis)*nis)*wi_cov
        px    = np.sqrt(xMean.T@_sigma@xMean)
        results.append(px)
    return np.argmin(np.array(results))
def get_accuracy(y,y_pred):
    return (y==y_pred).astype(int).sum()/y.shape[0]
```

## Funcion discriminante cuadratico

```python
def fdc_px(xi,MEANS,COVS,**kwargs):
    results= []
    xi = xi.reshape(-1,1)
    for wi_mean,wi_cov in zip(MEANS,COVS):
        sigmaIv = LA.inv(wi_cov)
        Wi      = -0.5* sigmaIv
        wi      =  sigmaIv@wi_mean
        w0i     = (-0.5*wi_mean@sigmaIv@wi_mean*-.5*np.log(LA.det(wi_cov))) + np.log(0.5)
        gx      = xi.T@Wi@xi + wi.T@xi + w0i
        results.append(gx)
    return np.argmax(np.array(results))
```

```python
def test_bayesian_clf(clf):
    accuracies =[]
    for INDEX in range(FOLDS):
        X_test,y_test = Xys_test[INDEX]
        y_pred        = classify(INDEX,X_test,clf)
        accuracy      = get_accuracy(y_test,y_pred)
        accuracies.append(accuracy)
    return np.array(accuracies)
# test_bayesian_clf(bayesClassifier)
```

# Perceptron

```python
def cost_fn(X,w):
    return (-w.T@X).sum()
def GD(X,**kwargs):
#     w             = np.random.rand(X.shape[0]).reshape(-1,1)
    wp=w          = np.zeros(X.shape[0]).reshape(-1,1)
    lr            = kwargs.get('learning_rate',0.01)
    epochs        = kwargs.get('epochs',100)
    show_progress = kwargs.get('show_progress',False )
    jpws          = []
    i             = 0
    predicate     = True
    max_iteration_reached = False
    convergence_reached = False
    predicate = True 
    h=hp=0 
    while(predicate):
        gxs                   = w.T@X
        mask0                 = (gxs >0).T.flatten()
        mask1                 = (gxs <=0).T.flatten()
        _X_                   = X[:,mask1]
        delta_Jw              = -(_X_.sum(axis=1)).reshape(-1,1)
        Jw                    = cost_fn(_X_,w)
        w                     = w + (lr*delta_Jw)
#       Pocket  
        h                    = mask0.astype(int).sum()     
        #print(h,hp)
        if(h>hp):
            wp = w
            hp=h
#       End pocket      
        max_iteration_reached = i>= epochs
        convergence_reached   = np.all(mask0)
        predicate             = not ( max_iteration_reached or convergence_reached )
        i+=1
    return w,convergence_reached,max_iteration_reached,wp,hp
    
def test_perceptron(**kwargs):
    epochs = kwargs.get("epochs",100)
    LR     = kwargs.get("learning_rate",0.0001)
    accurracies =[]
    acc2 =[]
    for INDEX in range(FOLDS):
        X_train,y_train = Xys_train[INDEX]
        _SS,(w0,w1)     = getWiByClass(X_train,y_train)
        XXX             = augmentX(np.concatenate([w0,-w1]).T)
        W,cr,mr,wp,hp   = GD(XXX,epochs=epochs,learning_rate=LR)
        X_test , y_test = Xys_test[INDEX]
        X               = augmentX(_SS.transform(X_test).T)
        n               = y_test.shape[0]
        res             = (((W.T@X)>0).astype(int) == y_test).astype(int).sum()/n
        res2            = (((wp.T@X)>0).astype(int) == y_test).astype(int).sum()/n
#         print("Fold[{}] Converge reached= {}\nMax. Iterations reached = {}\nhp={}\nn={}".format(INDEX,cr,mr,hp,n))
#         print("_"*100)
        accurracies.append(res)
        acc2.append(res2)
    return np.array(accurracies),np.array(acc2)
```

```python
perceptron_results,perceptron_results2  = test_perceptron(epochs=100,learning_rate=0.1)
perceptron_results.mean(),perceptron_results2.mean()
```

# Regresion logistica

#### Funcion sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

### Funcion de costo

```python
def cross_entropy_loss(w,yi,_yi,**kwargs):
    l = kwargs.get("l",0.001)
#     return -(yi*np.log(_yi)+(1-yi)*np.log(1-_yi)+((l/2) *LA.norm(w)**2)).sum()
#     return -(yi*np.log(_yi)+(1-yi)*np.log(1-_yi)).sum()+((l/2) *LA.norm(w)**2)
    return -(yi*np.log(_yi)+(1-yi)*np.log(1-_yi)).sum()+((l/2) *LA.norm(w)**2)
```

###  Descenso del gradiente

```python
def gradient_descent(X,y,**kwargs):
    LAMBDA       = kwargs.get("LAMBDA",1/(10**3))
    epochs       = kwargs.get("epochs",100)
    LR           = kwargs.get("learning_rate",0.001)
    threshold    = kwargs.get("threshold",1/(np.power(10,3)))
    show_entropy = kwargs.get("show_entropy",False)
    w            = np.zeros([D+1,1]).reshape(-1,1)
    predicate    =  True
    i            = 0 
    
    while(predicate):
#       Linear discriminant function
        x                   = (w.T@X)
#       Mapping values to (0,1)
        _y                  = sigmoid(x)
#       Supervised learning  
        yy                  = (_y-y)
#       Gradient 
        delta_Lw            = X@(yy.T)
#       w(i) vector of weights at iteration k
        w_temp              = w 
#       w(i+1) new vector of weights
        w                   = w-(LR*delta_Lw)  + LAMBDA*w
#       cross - entropy 
        Lw                  = cross_entropy_loss(w,y,_y,l=LAMBDA)
        if(show_entropy):
            print("[{}]- CROSS ENTROPY LOSS: {}".format(i,Lw))
#       check exit conditions: Max iteration reached or Convergence
        norm_diff           = LA.norm(np.abs(w_temp-w))
        max_iter_reached    = epochs-1<=i
        convergence_reached = norm_diff < threshold
        predicate = not(( max_iter_reached or convergence_reached ))
#       Increment the counter  
        i+=1 
    return {'w':w,
            'max_iter_reached':max_iter_reached,
            'convergence_reached':convergence_reached,
            "iterations":i
           }
```

```python
def test_logistic():
    accuracies = []
    for INDEX in range(FOLDS):
        st = StandardScaler()
        X_train,y_train = Xys_train[INDEX]
        st.fit(X_train)
        XXX             = augmentX( st.transform(X_train).T)
#         XXX             = augmentX( X_train.T)
        result          = gradient_descent(XXX,y_train,show_entropy=False,LAMBDA=0)
        w = result['w']
#         print("Fold[{}]\nMax. iter reached = {}\nConvergence reached={}".format(INDEX,mr,cr))
        X_test , y_test = Xys_test[INDEX]
        X               = augmentX(st.transform(X_test).T)
#         X               = augmentX(X_test.T)
        n               = y_test.shape[0]
        res             = (((w.T@X)>0).astype(int) == y_test).astype(int).sum()/n
        accuracies.append(res)
    return np.array(accuracies)
```

```python
logistic_results = test_logistic()
logistic_results.mean()
```
